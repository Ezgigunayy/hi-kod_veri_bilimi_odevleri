```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("IMDb_Dataset.csv")
df.head()
```

```python
def bilgi_al(df):
    print(df.shape)
    print("*****************")
    print(df.info())
    print("*****************")
    print(df.size)
    print("*****************")
    print(df.describe().T)
    print("*****************")
    print(df.isnull().any())
    print("*****************")
    print(df.head())
    print("*****************")
    print(df.tail())
    print("*****************")
```

```python
bilgi_al(df)
```
