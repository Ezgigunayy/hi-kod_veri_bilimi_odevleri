```python
#ödev1
maas = int(input("Lütfen maas bilgilerinizi giriniz:"))

if maas <= 10000:
    kesinti = maas * 0.05
elif maas <= 25000:
    kesinti = maas * 0.10
elif maas <= 45000:
    kesinti = maas * 0.25
else:
    kesinti = maas * 0.30

yeni_maas = maas - kesinti
print("Güncel maasınız: ", yeni_maas)
```

```python
#Ödev3
kullanici_adi = input("Lütfen kullanıcı adınızı oluşturunuz: ")

while True:
    sifre = input("Lütfen şifrenizi oluşturunuz: ")
    if 5 <= len(sifre) <= 10:
        print("Hesabınız oluşturuldu.")
        break
    else:
        print("Lütfen girdiniz şifre 5 haneden az 10 haneden fazla olmasın!")
```

```python
#Ödev4
kullanici_adi = input("Lütfen kullanıcı adınızı giriniz: ")
tanimli_sifre = "123456"
hak = 3

while hak > 0:
    sifre = input("Lütfen şifrenizi giriniz: ")
    if sifre == tanimli_sifre:
        print("Giriş yapıldı.")
        break
    else:
        hak -= 1
        if hak > 0:
            print(f"Yanlış şifre girildi! Kalan hakkınız: ", hak)
        else:
            print("Üç yanlış deneme yaptınız. Program sonlandırılıyor.")
```
